```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly as py
import plotly.offline as iplot
import plotly.express as px
import cufflinks as cf
%matplotlib inline
py.offline.init_notebook_mode(connected=True)
cf.go_offline()
cf.set_config_file(theme='pearl')
```

```python
df=pd.read_csv("admission - admission.csv")
#looking for sample data which taken randomely
df.sample(10)
```

```python
sns.barplot(x='University Rating', y='Chance of Admit', data=df)
```

BAR PLOT (using plotly)

```python
df_dup=df[['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research','Chance of Admit']]
for i in df_dup:
    agg_uni=df.groupby(i,as_index=False)['Chance of Admit'].mean()
    if i != 'Chance of Admit':
        fig=px.bar(agg_uni,x=i,y='Chance of Admit',title=i+' VS COA',text_auto=True)
        fig.show()
    else:
        pass
```

Scatter Plot

```python
df_dup=df[['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research','Chance of Admit']]
for i in df_dup:
    if i != 'Chance of Admit':
        fig=px.scatter(df_dup,x=i,y='Chance of Admit',title=i+' VS COA',template='presentation')
        fig.show()
    else:
        pass
```

Box plot:
        small points shows outlier in dataset

```python
df_dup=df[['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research','Chance of Admit']]
for i in df_dup:
    if i != 'Chance of Admit':
        fig=px.box(df_dup,x=i,y='Chance of Admit',title=i+' VS COA',template='ggplot2')
        fig.show()
    else:
        pass
```

Pie Chart (for categorical variables)

```python
df_dup=df[['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research','Chance of Admit']]
for i in df_dup:
    if i != 'Chance of Admit':
        if df[i].dtype=='object' or len(df[i].unique())<=10:
            agg_uni=df.groupby(i,as_index=False)['Chance of Admit'].mean()
            fig=px.pie(agg_uni,names=i,values='Chance of Admit',title=i+" vs COA",template='plotly_dark')
            fig.show()
    else:
        pass
```

Line Chart

```python
df_dup=df[['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research','Chance of Admit']]
for i in df_dup:
    if i != 'Chance of Admit':
        fig=px.line(df_dup,x=i,y='Chance of Admit',title=i+' VS COA',template='plotly_dark')
        fig.show()
    else:
        pass
```
