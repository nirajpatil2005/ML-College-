```python
import numpy as np
import pandas as pd
```

```python
ds = pd.read_csv("/content/temperatures.csv")
```

```python
X = ds["YEAR"].values[:, None]
y = ds["JUN"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
print("Mean absolute error: ", metrics.mean_absolute_error(y_test, y_pred))
print("Mean squared error: ", metrics.mean_squared_error(y_test, y_pred))
print("R2 score: ", metrics.r2_score(y_test, y_pred))
```

```python
plt.scatter(X_train, y_train, color="green", label = "Actual")
plt.plot(X_test, y_pred, color = "orange", label = "Predicted")
plt.xlabel("JUNE temperature")
plt.ylabel("Annual temperature")
plt.show()
```
